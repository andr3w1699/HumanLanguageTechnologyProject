# Use a Pretrained Model for Sentiment Analysy

Let's try to use a model like 'nlptown/bert-base-multilingual-uncased-sentiment', which is trained for 1–5 star ratings:

```python
from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Example usage
example_review = "This product is amazing! Totally worth the price."
result = sentiment_pipeline(example_review)

print(result)
```

```python
from sklearn.model_selection import train_test_split

# Assuming your dataframe is called df
train_df, val_df = train_test_split(df_train, test_size=0.2, random_state=42, stratify=df_train['label'])
```

```python
from transformers import pipeline

# Load sentiment analysis model trained to predict ratings from 1 to 5
sentiment_pipeline = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
!pip install -q tqdm
```

```python
def get_full_review(row):
    return f"{row['title']} {row['text']}"

# Create a new column with full review text
val_df['full_review'] = val_df.apply(get_full_review, axis=1)
```

```python
val_df.head()
```

```python
from tqdm.notebook import tqdm
# Split into manageable chunks (e.g., batches of 16)
batch_size = 32
texts = val_df['full_review'].tolist()
predictions = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]
    preds = sentiment_pipeline(batch)
    predictions.extend(preds)
```

```python
import re

# Extract numerical labels
predicted_labels = [int(re.search(r'\d+', pred['label']).group()) for pred in predictions]

# True labels
true_labels = val_df['label'].tolist()

# Evaluation
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(true_labels, predicted_labels))
print("Detailed Report:\n", classification_report(true_labels, predicted_labels))
```
